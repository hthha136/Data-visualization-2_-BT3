```python
df = pd.read_csv('shopeep_koreantop_clothing_shop_data.csv', encoding="latin1")
df.head(3)
```

```python
df['Hour']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.hour
df['Minute']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.minute
df['Second']=pd.to_datetime(df['response_time'],format=' %H:%M:%S').dt.second
df.head()
```

```python
x1 = df['response_rate']
x2 = df['Hour']
y1 = df['rating_good']

from matplotlib.pyplot import figure
fig = plt.figure(figsize=(12, 6))

ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)
#fig, ax = plt.subplots(1, 2)

ax1.scatter(x1, y1)
ax2.scatter(x2, y1)

ax1.set_ylabel('rating_good', fontsize = 14)
ax2.set_ylabel('rating_good', fontsize = 14)
ax1.set_xlabel('response_rate', fontsize = 14)
ax2.set_xlabel('response_time', fontsize = 14)

plt.show()
```
